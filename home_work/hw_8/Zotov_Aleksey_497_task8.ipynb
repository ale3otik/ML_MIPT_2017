```python
import numpy as np
import pylab as plt
%matplotlib inline
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    Sigma = 1.0 / l * F.dot(np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(Sigma)

    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / dTot * 100.0 for i in range(n)])
    m = sum(var_i < varRetained * 100)
    print('{:.2f}% variance retained in {:d} dimensions'.format(var_i[m], m))

    # plot the variance plot
    if show:
        plt.figure(figsize=(12,8))
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = V.T[:, :m]
    G = F.T.dot(U_reduced)

    return G, U_reduced
```

```python
plt.figure(figsize=(10,7))
colors = plt.cm.Spectral(np.arange(10)/9.0)
for i in range(10):
    X_i = X_projection[target==i]
    plt.scatter(X_i[:, 0], X_i[:,1],color=colors[i],label=i)
plt.title('data TSNE applied',fontsize=15)
plt.legend(loc = "best")
plt.show()
```

```python
tsne = TSNE(n_components=2)
features_tsne = tsne.fit_transform(features)
```

```python
def run_dbscan(eps,min_samples,metric):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit(features_tsne)
#     print(dbscan.labels_)
    return metric(dbscan.labels_,target)

def find_best_params(metric):
    best_res, best_eps, best_msmp = 0.0, 0.0, 0.0
    for eps in np.arange(1, 2, 0.01):
        for min_samples in np.arange(3, 14):
            cur_res = run_dbscan(eps, min_samples, metric)
            if cur_res > best_res:
                best_res, best_eps, best_msmp = cur_res, eps,min_samples

    print("Best params: Res:{:.3f}, eps={:.2f}, min_smp={:.2f}".format(best_res,best_eps,best_msmp))
find_best_params(metrics.adjusted_rand_score)
```

```python
find_best_params(metrics.adjusted_mutual_info_score)
```

Наилучшие параметры совпадают на обоих метриках

```python
best_eps = 1.96
best_min_smp = 13
dbscan = DBSCAN(eps=best_eps, min_samples=best_min_smp).fit(features_tsne)
nclusters = len(np.unique(dbscan.labels_))
colors = plt.cm.Spectral(np.linspace(0, 1, nclusters))
plt.figure(figsize=(10, 7))
for i, label in enumerate(np.unique(dbscan.labels_)):
    X_i = features_tsne[dbscan.labels_ == label]
    color = colors[i]
    if i == 0:
        color = 'black'
    plt.scatter(X_i[:, 0], X_i[:, 1], color=color, label=label)
plt.legend(loc='best')
plt.show()
```

```python
## CORE POINTS
n_cores = len(dbscan.core_sample_indices_)
print('n_cores: {}'.format(n_cores))
plt.figure(figsize=(15, 20))
for i in range(100):
    plt.subplot(10, 10, i + 1)
    plt.imshow(data.images[dbscan.core_sample_indices_[i]], cmap='binary') 
plt.show()
```

```python
## GARBAGE_POINTS
g_indices = [i for i, x in enumerate(dbscan.labels_) if x == -1]
print('n_garbage points: {}'.format(len(g_indices)))
plt.figure(figsize=(15, 20))
for i, x in enumerate(g_indices):
    plt.subplot(10, 10, i + 1)
    plt.imshow(data.images[x], cmap='binary') 
plt.show()
```
